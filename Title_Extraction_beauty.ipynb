```
dataset = pd.read_csv('beauty_data_info_train_competition.csv')
```

```
#filling nan value in benefits, colour group, and skin_type
dataset['Benefits'].fillna(7, inplace=True)
```

```
dataset['Colour_group'].value_counts()
```

```
dataset['Colour_group'].fillna(50, inplace = True)
```

```
dataset['Skin_type'].value_counts()
```

```
dataset['Skin_type'].fillna(10, inplace = True)
```

```
#brand nan nya sedikit, bisa dihilangkan saja
dataset['Brand'].value_counts()
data_brand = dataset.iloc[:,[1,4]]
data_brand = data_brand.dropna()
```

```
data_brand.head()
```

```
#x variable
title_data = dataset['title']
vectorizer = TfidfVectorizer()
vectorized = vectorizer.fit_transform(title_data)
```

```
x_brand = vectorizer.transform(data_brand['title'])
```

```
#product_texture nan nya sedikit, bisa dihilangkan saja
data_texture = dataset.iloc[:,[1,6]]
data_texture = data_texture.dropna()
```

```
data_texture.head()
```

```
x_texture = vectorizer.transform(data_texture['title'])
```

```
#Model for Benefit, Colour_Group, Skin_type

attributes = ['Benefits','Colour_group','Skin_type']

#model = naive bayes
from sklearn.naive_bayes import MultinomialNB
naive = MultinomialNB()
for i in attributes:
  x_train,x_test,y_train,y_test = train_test_split(vectorized, dataset[i], train_size = 0.2, random_state = 0)
  naive.fit(x_train,y_train)
  print('Attributes:', i)
  print('Train accuracy score : ', naive.score(x_train, y_train))
  print('Test Accuracy score:', naive.score(x_test, y_test))
```

```
#Model for Benefit, Colour_group, skin_type

attributes = ['Benefits','Colour_group','Skin_type']
list_forest_model = []
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier()
for i in attributes:
  forest = RandomForestClassifier()
  x_train,x_test,y_train,y_test = train_test_split(vectorized, dataset[i], train_size = 0.2, random_state = 0)
  forest.fit(x_train,y_train)
  list_forest_model.append(forest)
  print('model : forest_',i)
  print('Attributes:', i)
  print('Train accuracy score : ', forest.score(x_train, y_train))
  print('Test Accuracy score:', forest.score(x_test, y_test))
```

```
#iseng nyoba ngetrain pake logistic regression
#Model for Benefit, Colour_group, skin_type

attributes = ['Benefits','Colour_group','Skin_type']
from sklearn.linear_model import LogisticRegression
logs = LogisticRegression()
for i in attributes:
  x_train,x_test,y_train,y_test = train_test_split(vectorized, dataset[i], train_size = 0.2, random_state = 0)
  forest.fit(x_train,y_train)
  print('model : logistic regression_',i)
  print('Attributes:', i)
  print('Train accuracy score : ', forest.score(x_train, y_train))
  print('Test Accuracy score:', forest.score(x_test, y_test))
```

```
#urutan modelnya 'Benefits','Colour_group','Skin_type', Brand, Product Texture
print(list_forest_model)
```

```
#Separate model for Brand
from sklearn.naive_bayes import MultinomialNB
naive_Brand = MultinomialNB()
from sklearn.ensemble import RandomForestClassifier
forest_Brand = RandomForestClassifier()

model = [naive,forest]
for i in model:
  print(i)
  x_train,x_test,y_train,y_test = train_test_split(x_brand, data_brand['Brand'], train_size = 0.2, random_state = 0)
  i.fit(x_train,y_train)
  print('Attributes: Brand')
  if i == naive:
    print('model : naive_Brand')
    print('Attributes : Naive')
  elif i == forest:
    print('model :forest_Brand')
    print('Attributes : Forest')
    list_forest_model.append(i)
  print('Train accuracy score : ', i.score(x_train, y_train))
  print('Test Accuracy score:', i.score(x_test, y_test))
```

```
#Separate model for Texture
from sklearn.naive_bayes import MultinomialNB
naive = MultinomialNB()
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier()

model = [naive,forest]
for i in model:
  print(i)
  x_train,x_test,y_train,y_test = train_test_split(x_texture, data_texture['Product_texture'], train_size = 0.2, random_state = 0)
  i.fit(x_train,y_train)
  print('Attributes: Brand')
  if i == naive:
    print('Model : Naive')
  elif i == forest:
    print('Model : Forest')
    list_forest_model.append(i)
  print('Train accuracy score : ', i.score(x_train, y_train))
  print('Test Accuracy score:', i.score(x_test, y_test))
```
